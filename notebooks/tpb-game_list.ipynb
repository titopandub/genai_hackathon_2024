```python
import pandas as pd

df = pd.read_csv("data/game_list.csv")
df.dropna(inplace=True)
df
```

```python
def generate_markdown_from_df(df):
    games_text = []
    games_text.append("**List Vidio.com Arcade Games**\n\n[Link Bermain](https://m.vidio.com/games)")
    for name, url, description, thumbnail in df.values.tolist():
        description_cleaned = description.replace('\n\n', '\n')
        games_text.append(f"- {name}\n\n  {description_cleaned}\n\n  [![{name}]({thumbnail})]({url})")
    return "\n\n".join(games_text)
```

```python
from IPython.display import display, Markdown

Markdown(generate_markdown_from_df(df))
```

```python
from datetime import datetime
from ddtrace import tracer
from langchain_google_vertexai import VertexAI
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate

model = VertexAI(model_name="gemini-pro", project="vidio-quiz-prod", location="asia-southeast1")

def vidio_info_chain(context, user_query):
    chain = (ChatPromptTemplate.from_template(
"""
You are chatbot for Vidio OTT platform. Your task is to answer to user question.
You are a chatbot for the Vidio OTT platform. Today's date is {today}. When responding to user questions, provide information, schedules, or match details only if they are scheduled for a future date relative to today. If all provided information pertains to past dates, respond with "I don't have the current information." Avoid giving any outdated information about schedules or matches.

Questions can be related to Vidio products, games, films, or other topics.

For queries specifically about games or arcade games available on Vidio, limit your response to three relevant game options and include direct links for users to play these games.

Here are the relevant pieces of information:
{information} 

User Question:
{user_query}
"""
    ) | model | StrOutputParser())
    information = generate_markdown_from_df(df)
    today_date = datetime.now().strftime("%d %B %Y")

    resp = chain.invoke({"user_query": user_query, "information": information, "today": today_date})

    return resp
```

```python
response = vidio_info_chain(None, "game apa yang seru dimainkan saat menunggu antrian?")
```

```python
Markdown(response)
```
